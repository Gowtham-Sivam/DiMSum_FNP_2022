```python
keywords_weights = {}
stopwords = list(STOP_WORDS)
for index, row in greek_toc_section_df.iterrows():
    nlp_doc = nlp(row['greek'])
    print(nlp_doc)
    for token in nlp_doc:
        if(token.text in stopwords or token.text in punctuation):
            continue
        try:
            keywords_weights[token.text] = keywords_weights[token.text] + row['score']
        except:
            keywords_weights[token.text] = row['score']
```

```python
keywords_weights
```

```python
keywords_weights = {k: v for k, v in sorted(keywords_weights.items(), key=lambda item: item[1], reverse=True)}
```

```python
def get_summary(text, num_sents = 3):
    nlp_text_doc = nlp(text)
    sent_strength={}
    for sent in nlp_text_doc.sents:
        for word in sent:
            if word.text in keywords_weights.keys():
                if sent in sent_strength.keys():
                    sent_strength[sent]+=keywords_weights[word.text]
                else:
                    sent_strength[sent]=keywords_weights[word.text]
    summarized_sentences = nlargest(num_sents, sent_strength, key=sent_strength.get)
    final_sentences = [ w.text for w in summarized_sentences ]
    summary = ' '.join(final_sentences)
    return summary
```
